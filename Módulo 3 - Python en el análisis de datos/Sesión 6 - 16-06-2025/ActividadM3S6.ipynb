```python
# Inventario nivel intermedio
inventario_intermedio = {
    "pan": {"stock": 10, "precio": 1200, "proveedores": ["Panadería Don Pan", "Superpan"]},
    "leche": {"stock": 0, "precio": 950, "proveedores": ["Lechera Norte"]},
    "huevos": {"stock": 30, "precio": 180, "proveedores": ["Granja del Sur", "Huevos Spa"]},
    "café": {"stock": 5, "precio": 2500, "proveedores": ["Distribuidora Aroma"]},
    "queso": {"stock": 15, "precio": 2800, "proveedores": ["Quesería Andes"]},
    "mantequilla": {"stock": 0, "precio": 5000, "proveedores": ["Granja del Sur", "Quesería Andes"]},
}

#print(type(inventario_intermedio))

#Codigo

#1. Mostrar todos los productos con sus cantidades.
def mostrar_stock(inventario):
    print("\n---STOCK DEL INVENTARIO---")
    print(f"{'Producto'.ljust(15)}Stock")
    print("-" * 25)
    for objeto, detalles in inventario.items():
        print(f"{objeto.ljust(15)}{str(detalles['stock'])}")
    input("Apreta enter para continuar...")


#2. Verificar si hay productos agotados (cantidad = 0) y permitir actualizarlos.
def productos_agotados(inventario):
    while True:
        print("\n---CONSULTA PRODUCTOS AGOTADOS---")
        lista_agotados = []
        for objeto, detalles in inventario.items():
            if detalles['stock']==0:
                lista_agotados.append(objeto)
                print(f'{objeto} tiene {detalles["stock"]} stock')
        if len(lista_agotados)>0:
            try:
                opcion = int(input("\n---¿Que desea hacer?---\n1)Modificar cantidad\n2)Salir\nSeleccione una opción: "))
                if opcion == 1:
                    cambiar_producto(inventario,lista_agotados)
                elif opcion == 2:
                    return
                else:
                    print("\nERROR: Ingrese una opción de la lista")
            except ValueError: 
                print("\nERROR: Ingrese un numero entero")
        else:
            print("No se encuentran mas inventario sin stock, volviendo al menu")
            input("Apreta enter para continuar...")
            return

def cambiar_producto(inventario, agotados):
    while True:
        print("\n---Ingrese el objeto quiere modificar---")
        for i, producto in enumerate(agotados, 1):
            print(f"{i}) {producto}")
        print(f"{len(agotados)+1}) Salir")
        try:
            opcion=int(input("Seleccione una opción: "))
            if 0<opcion<=len(agotados):
                while True:
                    print(f"\n---Modificando {agotados[opcion-1]:}")
                    try:
                        cantidad=int(input("Ingrese la nueva cantidad mayor a 0: "))
                        if cantidad>0:
                            inventario[agotados[opcion-1]]["stock"]=cantidad
                            print(f"¡Se ha cambiado con exito la cantidad de {agotados[opcion-1]}!")
                            input("Apreta enter para continuar...")
                            #mostrar_productos(inventario)
                            return
                        else:
                            print("\nERROR: Ingrese una cantidad mayor a 0")
                    except ValueError:
                        print("\nERROR: Ingrese un número entero")
            elif opcion==len(agotados)+1:
                break
            else:
                print("\nERROR: Ingrese un número de la lista")
        except ValueError:
            print("\nERROR: Ingrese un número entero")

#3. Permitir al usuario agregar nuevos productos al inventario.
def agregar_producto(inventario):
    print("\n---AGREGAR NUEVO PRODUCTO---")
    nombre = input("Ingrese el nombre del producto: ").strip().lower()
    if nombre in inventario:
        print("Ese producto ya existe en el inventario.")
        return
    while True:
        try:
            stock = int(input("Ingrese el stock inicial"))
            if stock < 0:
                print("ERROR: El stock no puede ser negativo.")
            else:
                break
        except ValueError:
            print("ERROR: Ingrese un número entero válido para el stock.")

    while True:
        try:
            precio = int(input("Ingrese el precio"))
            if precio < 0:
                print("ERROR: El precio no puede ser negativo.")
            else:
                break
        except ValueError:
                print("ERROR: Ingrese un número entero válido para el precio.")

    while True:
        try:
            cantidad_proveedores = int(input("Ingrese la cantidad de proveedores: "))
            if cantidad_proveedores <= 0:
                print("ERROR: Debe ingresar al menos un proveedor.")
            else:
                break
        except ValueError:
            print("ERROR: Ingrese un número entero válido.")

    proveedores = []
    for i in range(cantidad_proveedores):
        while True:
            proveedor = input(f"Ingrese el nombre del proveedor #{i+1}: ").strip()
            if proveedor == "":
                print("ERROR: El nombre no puede estar vacío.")
            elif proveedor in proveedores:
                print("ERROR: Este proveedor ya fue ingresado.")
            else:
                proveedores.append(proveedor)
                break

    inventario[nombre] = {
        "stock": stock,
        "precio": precio,
        "proveedores": proveedores
    }
    print(f"Producto '{nombre}' agregado con éxito.")
    input("Apreta enter para continuar...")

#4. Mostrar los productos con más de 10 unidades disponibles.
def mostrar_sobre_10(inventario):
    print("\n---INVENTARIO SOBRE 10 UNIDADES---")
    hay_productos = False
    for objeto, detalles in inventario.items():
        if detalles['stock'] >= 10:
            print(f"{objeto}: {detalles['stock']}")
            hay_productos = True
    if not hay_productos:
        print("No hay productos con más de 10 unidades.")
    input("Apreta enter para continuar...")

#5. Al finalizar, mostrar el resumen del inventario actualizado.
def mostrar_todo(inventario):
    print("\n---INVENTARIO DETALLADO---")
    print(f"{'Producto'.ljust(15)}{'Stock'.rjust(7)}{'Precio'.rjust(10)}    Proveedores")
    print("-" * 70)
    for producto, datos in inventario.items():
        nombre = producto.ljust(15)
        stock = str(datos["stock"]).rjust(7)
        precio = f"${datos['precio']}".rjust(10)
        proveedores = ", ".join(datos["proveedores"])
        print(f"{nombre}{stock}{precio}    {proveedores}")
    input("Apreta enter para continuar...")

def menu_principal(inventario):
    while True:
        print("\n--- MENÚ INVENTARIO ---")
        print("1) Mostrar productos")
        print("2) Verificar productos agotados")
        print("3) Agregar nuevo producto")
        print("4) Consultar productos sobre 10 unidades")
        print("5) Inventario detallado")
        print("6) Salir")
        try:
            opcion = int(input("Seleccione una opción: "))
            if opcion == 1:
                mostrar_stock(inventario)
            elif opcion == 2:
                productos_agotados(inventario)
            elif opcion == 3:
                agregar_producto(inventario)
            elif opcion == 4:
                mostrar_sobre_10(inventario)
            elif opcion == 5:
                mostrar_todo(inventario)
            elif opcion == 6:
                print("Saliendo del programa...")
                break
            else:
                print("Opción inválida.")
        except ValueError:
            print("Debe ingresar un número.")

menu_principal(inventario_intermedio)
```
