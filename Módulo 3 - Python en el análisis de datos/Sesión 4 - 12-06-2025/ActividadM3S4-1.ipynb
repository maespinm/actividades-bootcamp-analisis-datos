```python
### DATOS ###

clientes = ['Ana Rojas', 'Luis Pérez', 'Claudia Soto', 'Pedro Gómez', 'Sofía Núñez']
info_clientes = {
    'Ana Rojas': {'telefono': '912345678', 'correo': 'ana.rojas@techmail.com'},
    'Luis Pérez': {'telefono': '987654321', 'correo': 'luisperez@techmail.com'},
    'Claudia Soto': {'telefono': '965432198', 'correo': 'claudia.soto@techmail.com'},
    'Pedro Gómez': {'telefono': '923456789', 'correo': 'pedrogomez@techmail.com'},
    'Sofía Núñez': {'telefono': '933221144', 'correo': 'sofia.nunez@techmail.com'}
}
clientes_por_ciudad = {
    'Antofagasta': ['Ana Rojas', 'Luis Pérez'],
    'Santiago': ['Claudia Soto'],
    'Valparaíso': ['Pedro Gómez', 'Sofía Núñez']
}
servicios_clientes = {
    'Ana Rojas': {'Hosting', 'Diseño Web'},
    'Luis Pérez': {'Correo Empresarial', 'Consultoría TI'},
    'Claudia Soto': {'Diseño Web', 'SEO'},
    'Pedro Gómez': {'Hosting', 'Correo Empresarial'},
    'Sofía Núñez': {'Desarrollo App', 'Consultoría TI'}
}
```

```python
###
#1.- Buscar clientes por santiago
###
def buscar_por_ciudad(nombre):
    if nombre in clientes_por_ciudad:
        print(f"Los clientes de la ciudad {nombre} son {clientes_por_ciudad[nombre]}")
    else:
        print("No se encuentra la ciudad ingresada")

###
#2.- Agregar un nuevo cliente
###
def crear_cliente():
    nombre = input("Ingrese el nombre y apellido del cliente :")
    while True:
        try:
            telefono = int(input("Ingrese el numero de telefono del cliente :"))
            if (len(str(telefono))==9):
                break
            else:
                print("Error, ingrese un numero telefonico de 9 numeros")
        except ValueError:
            print("Error, ingrese un numero entero")
    correo=input("Ingrese su correo electronico")
    clientes.append(nombre)
    info_clientes[nombre]={'telefono':telefono,'correo':correo}

###
#3.- Agregar un servicio a un cliente sin duplicarlo
###

def crear_servicio(nombre_cliente):
    if nombre_cliente in servicios_clientes:
        servicio = input(f"Ingrese el nuevo servicio a agregar a: {nombre_cliente}")
        if servicio in servicios_clientes[nombre_cliente]:
            print("Error, el servicio ya existe en esta persona")
        else:
            servicios_clientes[nombre_cliente].add(servicio)
            print(f"Servicio '{servicio}' agregado a {nombre_cliente}.")
    else:
        print("La persona ingresada no existe")
    
###
#4.- Consultar la cantidad de clientes y/o servicio unicos
###
def cantidad_clientes_servicios():
    print("La cantidad de clientes es: ",len(clientes))
    print(servicios_clientes.values())
    servicios_unicos = set.union(*servicios_clientes.values())
    print("La cantidad de servicios unicos es: ",len(servicios_unicos))

###
#5.- Indicar cuántos clientes tienen "Hosting" contratado
###
def contar_servicio(nombre_servicio):
    clientes_con_servicio = list(filter(lambda cliente: nombre_servicio in servicios_clientes[cliente], servicios_clientes))
    if clientes_con_servicio:
        print(f"La cantidad de clientes con el servicio '{nombre_servicio}' es {len(clientes_con_servicio)}")
        print("Clientes:", clientes_con_servicio)
    else:
        print("El servicio no existe")


#buscar_por_ciudad("Santiago")
#crear_cliente()
#print(clientes)
#print(info_clientes)
#crear_servicio("Ana Rojas")
#print(servicios_clientes)
cantidad_clientes_servicios()
contar_servicio("Hosting")
#for cliente,datos in info_clientes.items(): print(cliente,datos)      
    
```
