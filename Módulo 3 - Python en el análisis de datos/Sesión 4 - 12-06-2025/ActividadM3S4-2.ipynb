```python
clientes = ['Ana Rojas', 'Luis Pérez', 'Claudia Soto', 'Pedro Gómez', 'Sofía Núñez']

info_clientes = {
    'Ana Rojas': {'telefono': '912345678', 'correo': 'ana.rojas@techmail.com'},
    'Luis Pérez': {'telefono': '987654321', 'correo': 'luisperez@techmail.com'},
    'Claudia Soto': {'telefono': '965432198', 'correo': 'claudia.soto@techmail.com'},
    'Pedro Gómez': {'telefono': '923456789', 'correo': 'pedrogomez@techmail.com'},
    'Sofía Núñez': {'telefono': '933221144', 'correo': 'sofia.nunez@techmail.com'}
}

clientes_por_ciudad = {
    'Antofagasta': ['Ana Rojas', 'Luis Pérez'],
    'Santiago': ['Claudia Soto'],
    'Valparaíso': ['Pedro Gómez', 'Sofía Núñez']
}

servicios_clientes = {
    'Ana Rojas': {'Hosting', 'Diseño Web'},
    'Luis Pérez': {'Correo Empresarial', 'Consultoría TI'},
    'Claudia Soto': {'Diseño Web', 'SEO'},
    'Pedro Gómez': {'Hosting', 'Correo Empresarial'},
    'Sofía Núñez': {'Desarrollo App', 'Consultoría TI'}
}

servicios_disponibles = ('Hosting', 'Diseño Web', 'Correo Empresarial', 'SEO', 'Consultoría TI', 'Desarrollo App')


###
# 1. Mostrar la cantidad total de clientes registrados.
####

def cantidad_clientes():
    return len(info_clientes)

###
# 2. Mostrar el correo electrónico de “Claudia Soto”.
####

def mostrar_correo(nombre_usuario):
    if nombre_usuario in info_clientes:
        correo = info_clientes[nombre_usuario]["correo"]
        return correo
    else:
        print("El usuario no existe")
        
###
# 3. Agregar un nuevo cliente llamado “Valeria Díaz” con los siguientes datos: Teléfono: 922334455, Correo: valeria.diaz@techmail.com, Ciudad: Santiago, Servicios: 'Diseño Web', 'Desarrollo App'
####

def crear_cliente(nombre,telefono,correo,ciudad,servicios):
    clientes.append(nombre)
    info_clientes[nombre]={'telefono':telefono,'correo':correo}
    if ciudad in clientes_por_ciudad:
        clientes_por_ciudad[ciudad].append(nombre)
    else:
        clientes_por_ciudad[ciudad]={nombre}
        servicios_clientes[nombre]=set(servicios)

#1-1
print("La cantidad de clientes es:",cantidad_clientes())
print(f"Clientes: {", ".join(clientes)}")

#1-2
print("\nEl correo de Claudia Soto es:",mostrar_correo("Claudia Soto"))

#1-3
crear_cliente("Valeria Díaz",922334455,"valeria.diaz@techmail.com","Concepcion",['Diseño Web','Desarrollo App'])
print("\nSe ha creado el usuario Valeria Díaz")
print(info_clientes["Valeria Díaz"])

#1-4
crear_cliente("Mark Espinoza",123456789,"maespinm@example.com","Talcahuano",['Diseño Web','Desarrollo App','Analisis de Datos'])
print("\nSe ha creado el usuario Mark Espinoza")
print(info_clientes["Mark Espinoza"])

#2-5
print("\nLos clientes de Valparaiso son:", ", ".join(clientes_por_ciudad["Valparaíso"]))

#3-6
nuevos = "Mario Paredes, Camila Ruiz, Tomás Díaz"
clientes.extend(nuevos.split(", "))
print(f"\nSe han agregado los usuario {nuevos} a clientes")
print(f"Clientes: {", ".join(clientes)}")

#3-7
calificaciones=[[5,8,3],[4,2,5],[9,10,10]]
print("\nCalificaciones:")
print(f"\t\n{clientes[0]}\nHosting: {calificaciones[0][0]}\nDiseño Web: {calificaciones[0][1]}\nCorreo empresarial: {calificaciones[0][2]}\n")
print(f"\t\n{clientes[1]}\nHosting: {calificaciones[1][0]}\nDiseño Web: {calificaciones[1][1]}\nCorreo empresarial: {calificaciones[1][2]}\n")
print(f"\t\n{clientes[2]}\nHosting: {calificaciones[2][0]}\nDiseño Web: {calificaciones[2][1]}\nCorreo empresarial: {calificaciones[2][2]}\n")
```
