```python
import numpy as np

precios = np.array([15000,22000,18000,25000,30000])
cantidad_unidades_vendidas=np.array([[5, 8, 6, 7, 10], [3, 6, 5, 4, 7], [8, 12, 10, 9, 11], [4, 5, 6, 3, 4], [6, 7, 8, 5, 9]])

#1.- Cálculo de ingresos diarios: Multiplica la matriz de unidades vendidas por el vector de precios de los productos para obtener los ingresos diarios de cada producto.
def ingreso_diario_producto(precios, cantidad):
    ingreso = precios * cantidad
    print("\n\n---INGRESO DIARIO POR PRODUCTO---")
    
    # Encabezado con productos
    encabezado = "Día".ljust(6)
    for i in range(ingreso.shape[1]):
        encabezado += f"Prod{i+1}".rjust(12)
    print(encabezado)
    print("-" * len(encabezado))

    # Mostrar ingresos alineados
    for i, fila in enumerate(ingreso):
        linea = f"{str(i+1).ljust(6)}"
        for valor in fila:
            linea += f"{valor:>12,}"  # Alineado a la derecha con separador de miles
        print(linea)
        
#2.- Ingresos totales por día: Suma los ingresos diarios para conocer la recaudación total de la tienda por cada día.
def ingreso_diario_total(precios, unidades):
    ingreso=precios*unidades
    print("\n\n---INGRESO DIARIO TOTAL---")
    for i, ingreso_dia in enumerate(ingreso):
        print(f"Dia {i+1}:  ${sum(ingreso_dia):,}")

#3.- Día con mayores ingresos: Determina cuál fue el día con la mayor recaudación y cuánto se vendió en total. 
def dia_mayor_ingreso(precios, unidades):
    ingreso=precios*unidades
    ingresos_dia = np.sum(ingreso, axis=1)
    dia_max = np.argmax(ingresos_dia)
    print("\n\n---EL DIA CON MAYOR INGRESO---")
    print(f"El dia con mayor ingreso fue el dia {dia_max+1} con una cantidad de ${np.max(ingresos_dia):,}")

#4.-Productos más rentables: Identifica cuáles fueron los dos productos que generaron más ingresos en la semana. 
def producto_rentable_semana(precios,unidades):
    print("\n\n---EL PRODUCTO MAS RENTABLE DE LA SEMANA---")
    ingreso=precios*unidades
    ingreso_producto=np.sum(ingreso, axis=0)
    print(f"El producto {np.argmax(ingreso_producto)+1} fue el mas rentable de la semana con un monto de ${np.max(ingreso_producto):,}")

#5.- Filtrado de ventas altas: Utiliza selección condicional para identificar los días en los que las ventas totales superaron los $400.000.
def ventas_superan(precios,cantidad,monto):
    print(f"\n\n---DIAS QUE SUPERAN EL MONTO DE ${monto:,}---")
    ingreso=precios*cantidad
    ingresos_dia = np.sum(ingreso, axis=1)
    for i, dia in enumerate(ingresos_dia):
        #print(np.sum(semana)) 
        if (np.sum(dia)>=monto):
            print(f"El dia {i+1} supero la cantidad de ${monto:,}")

ingreso_diario_producto(precios,cantidad_unidades_vendidas)
ingreso_diario_total(precios,cantidad_unidades_vendidas)
dia_mayor_ingreso(precios,cantidad_unidades_vendidas)
producto_rentable_semana(precios,cantidad_unidades_vendidas)
ventas_superan(precios,cantidad_unidades_vendidas,400000)
```
